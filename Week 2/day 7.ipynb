# Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of analyzing and summarizing datasets to uncover patterns, relationships, and insights. It helps in understanding the structure of the data, identifying anomalies, and preparing it for further analysis or modeling.

### Key Steps in EDA:
1. **Understanding the Dataset**:
    - Inspect the dataset structure (rows, columns, data types).
    - Check for missing values and duplicates.

2. **Descriptive Statistics**:
    - Summarize data using measures like mean, median, mode, standard deviation, etc.
    - Analyze distributions of numerical and categorical variables.

3. **Data Visualization**:
    - Use plots like histograms, box plots, scatter plots, and heatmaps to visualize data distributions and relationships.

4. **Correlation Analysis**:
    - Identify relationships between variables using correlation coefficients or pair plots.

5. **Handling Missing or Outlier Data**:
    - Detect and decide how to handle missing values or outliers.

EDA is a crucial step in the data analysis pipeline, ensuring better decision-making and model performance.

```python
# task 1: data cleaning aggregation and fitlering.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load titanic dataset
url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
# print(df.info())
# print(df.describe())

# handle missing values age, cabin, embarked

df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Cabin"] = df["Cabin"].fillna("Unknown")
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# remove duplicates
df = df.drop_duplicates()
df.info()

# filter dataset
first_class = df[df["Pclass"] == 1]
print(f"Fist class passengers: {first_class.head()}")

# bar chart: survival by class
survival_by_class = df.groupby("Pclass")["Survived"].mean()
survival_by_class.plot(kind="bar")
plt.title("Survival rate by class")
plt.xlabel("Class")
plt.ylabel("Survival rate")
plt.show()

# histogram: age distribution
sns.histplot(df["Age"], kde=True, bins=20, color="purple")
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# scatter plot: age vs fare
plt.scatter(df["Age"], df["Fare"], alpha=0.5, color="green")
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
