```python
import matplotlib.pyplot as plt
```

```python
# Basic Plot
x = [1, 2,3,4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()
```

```python
# Line Plot

plt.plot([1,2,3], [10, 20, 30], label='Trend')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Bar Chart
categories = ["A", "B", "C"]
values = [10, 15, 7]
plt.bar(categories, values, color='blue')
plt.show()
```

```python
# histogram

data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4, color='green', edgecolor='black')
plt.show()
```

```python
# scatter plot
x = [ 1, 2, 3, 4, 5]
y = [ 10, 12, 25, 30, 45]
plt.scatter(x, y, color='red')
plt.title('Scatter Plot')
plt.show()
```

```python
# Excercise 1: create basic plot with matlaplotlib
import matplotlib.pyplot as plt
# Line Plot
years = [2010, 2011, 2012, 2013]
sales = [ 100, 120, 140, 160]
plt.plot(years, sales, label="Sales Trend", color='blue', marker='o')
plt.title("Sales Over Years")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.legend()
plt.show()

```

```python
# Bar Chart
categories = ["electronics", "clothing", "groceries"]
revenue = [250, 400, 150]
plt.bar(categories, revenue, color='green')
plt.title("Revenue by Category")
plt.xlabel("Categories")
plt.ylabel("Revenue")
plt.show()
```

```python

# scatter plot

hours_studied = [1, 2, 3, 4, 5]
exam_scores = [50, 55, 65, 70, 85]
plt.scatter(hours_studied, exam_scores, color='red')
plt.title('Scatter Plot')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.show()
```
