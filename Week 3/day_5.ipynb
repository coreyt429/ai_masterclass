# Probability Basics
- Conditional Probability
  The probability of an event A given that event B has occurred
  $P(A | B) = \frac{P(A \cap B)}{P(B)}$
- Bayes' Theorem
  $P(A | B) = \frac{P(A | B)P(A)}{P(B)}$
    - P(A) Prior probability
    - P(B|A) Likelihood
    - P(B) Evidence
  

```python
def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate the posterior probability using Bayes' theorem.

    Parameters:
    prior (float): The prior probability P(A).
    likelihood (float): The likelihood P(B|A).
    evidence (float): The evidence P(B).

    Returns:
    float: The posterior probability P(A|B).
    """
    return (prior * likelihood) / evidence if evidence != 0 else 0
# Example usage
```

# Common Probability Distributions
- Gaussian (Normal) Distribution
    - $\displaystyle f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$
- Binomial Distribution
    - Models the number of successes in n independent Bernoulli trials
    - $\displaystyle P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$
- Poisson Distrubution
    - Models the number of events in a fixed interval of time or space
    - $\displaystyle P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$

```python
# hands on excercises

# 1: Calculate Probabilities Using Bayes' Theorem
# Problem:
# - A disease affects 1% of the population.
# - A test is 95% accurate for diseeased individuals (true positive rate).
# - The test is 90% accurate for healthy individuals (true negative rate).
# - What is the probability that a person has the disease given a positive test result?
# Given:
# - P(Disease) = 0.01 (prior probability)
# - P(Positive Test | Disease) = 0.95 (likelihood)
# - P(Positive Test | Healthy) = 0.10 (false positive rate)
# - P(Healthy) = 1 - P(Disease) = 0.99 (prior probability of being healthy)
# - P(Positive Test) = P(Positive Test | Disease) * P(Disease) + P(Positive Test | Healthy) * P(Healthy)
# - P(Positive Test) = 0.95 * 0.01 + 0.10 * 0.99
# Solution:
def bayes_theorem(prior, sensitivity, specificity):
    """
    Calculate the posterior probability using Bayes' theorem.
    Parameters:
    prior (float): The prior probability P(Disease).
    sensitivity (float): The true positive rate P(Positive Test | Disease).
    specificity (float): The true negative rate P(Positive Test | Healthy).
    Returns:
    float: The posterior probability P(Disease | Positive Test).
    """
    evidence = (sensitivity * prior) + ((1 - specificity) * (1 - prior))
    posterior = (sensitivity * prior) / evidence if evidence != 0 else 0
    return posterior

prior = 0.01
sensitivity = 0.95
specificity = 0.90
posterior = bayes_theorem(prior, sensitivity, specificity)
print(f"Posterior Probability of Disease given Positive Test: {posterior:.4f}")

# 2: 
```

```python
# Excercise 2: Plot ans Explore Different Probability Distributions
# Problem:
# - Explore and visualize different probability distributions using Python.
# - Use the following distributions: Normal, Bernoulli, Binomial, and Poisson.
# - For each distribution, generate random samples and plot the probability mass function (PMF) or probability density function (PDF).
# Solution:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli, binom, poisson
# Gaussian (Normal) Distribution
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, loc=0, scale=1)
plt.plot(x, y, label='Gaussian Distribution')
plt.title('Gaussian Distribution')
plt.show()
```

```python
# binomial distribution
n, p = 10, 0.5
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)
plt.bar(x, y, label='Binomial Distribution', color='blue')
plt.title('Binomial Distribution')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.show()
```

```python
# Bernoulli Distribution
n = 1
p = 0.5
x = np.arange(0, n+1)
y = bernoulli.pmf(x, p)
plt.bar(x, y, label='Bernoulli Distribution', color='green')
plt.title('Bernoulli Distribution')
plt.xlabel('Success (1) or Failure (0)')
plt.ylabel('Probability')
plt.show()
```

```python
# poisson distribution
lam = 3
x = np.arange(0, 10)
y = poisson.pmf(x, lam)
plt.bar(x, y, label='Poisson Distribution', color='orange')
plt.title('Poisson Distribution')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.show()

```
